```python
%matplotlib inline

import numpy as np
import matplotlib
import matplotlib.pyplot as plt

n = np.arange(0, 4000, 1)

plt.plot(n, n, label='n')
plt.plot(n, 2 * n + 10, label='2 * n + 10')
plt.plot(n, 3 * n + 20, label='3 * n + 20')
plt.title('Complexity Functions')
plt.xlabel('n')
plt.ylabel('T(n)')
plt.legend(loc=2)

# double the resolution of the figure
config = plt.gcf()
size = config.get_size_inches()
config.set_size_inches((size[0]*2, size[1]*2))
matplotlib.rcParams.update({'font.size': 22})

plt.tight_layout()
plt.savefig('complexity1.png')

plt.show()
```
